```python
# Calculate RUL for each unit
rul_colum_names = ["RUL"]
rul_data_path = r"C:\Users\risha\Program_folder\Projects\Demo_projects\CMAPSSData\RUL_FD001.txt"
rul_data = pd.read_csv(rul_data_path, sep='\s+', header=None, names=rul_colum_names)
rul_data["unit_number"] = train_data["unit_number"].unique()
rul_data = rul_data[["unit_number", "RUL"]]
rul_data.head()
```

```python
rul_data.shape
```

```python
for sensor in [f"sensor_reading_{i}" for i in range(1,22)]:
    train_data[f'{sensor}_rolling_mean'] = train_data.groupby('unit_number')[sensor].rolling(window=3).mean().reset_index(level=0, drop=True)

train_data.head()
```

```python
train_data = train_data.merge(rul_data, on=['unit_number'], how='left')
train_data['RUL_val'] = train_data['RUL'] - train_data['time_in_cycles']
train_data.drop(columns=['RUL'], inplace=True)

train_data.head()
```

```python
train_data.isna().sum()
```

```python
train_data.fillna(method="ffill", inplace=True)
train_data.head()
```

```python
train_data = train_data.dropna()
train_data.head()
```

```python
train_data.columns
```

```python
train_data.describe()
```

```python
columns_names = train_data.columns
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

columns_to_normalize = columns_names[2:-1]
train_data[columns_to_normalize] = scaler.fit_transform(train_data[columns_to_normalize])

train_data.head(10)
```

```python
from sklearn.model_selection import train_test_split
feature_columns = [col for col in train_data.columns if col not in ["unit_number", "time_in_cycles", "RUL"]]

X = train_data[feature_columns]
Y = train_data["RUL_val"]

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor

# Initialize the model
model = RandomForestRegressor()

# Train the model
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Predict on the test set
y_pred = model.predict(x_test)

# Calculate evaluation metrics
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
R2_score = r2_score(y_test, y_pred)

mse, mae, R2_score
```

```python
model_score = model.score(x_train, y_train)
print(f'R^2 Score: {model_score}')
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gb = GradientBoostingRegressor()

gb.fit(x_train, y_train)
```
